## Part 1

```python
import numpy as np
from collections import defaultdict

input_file_path = "test_input.txt"
input_data = open(input_file_path).read().strip()
input_lines = input_data.split('\n')
grid = np.array([[c for c in line] for line in input_lines])

rows, cols = grid.shape

sum_of_numbers = 0
nums = defaultdict(list)

for row_num in range(rows):
    gears = set()
    current_number_str = ""
    adj = False

    for col_num in range(cols + 1):
        if col_num < cols and grid[row_num, col_num].isdigit():
            current_number_str += grid[row_num, col_num]

            neighbors = grid[max(0, row_num - 1):min(rows, row_num + 2),
                             max(0, col_num - 1):min(cols, col_num + 2)]

            for ch in neighbors.flatten():
                if not ch.isdigit() and ch != '.':
                    adj = True
                if ch == '*':
                    gears.add((row_num, col_num))
        elif current_number_str:
            current_number = int(current_number_str)
            for gear in gears:
                nums[gear].append(current_number)
            if adj:
                sum_of_numbers += current_number
            current_number_str = ""
            adj = False
            gears = set()

print(dict(nums))
print(sum_of_numbers)
```
