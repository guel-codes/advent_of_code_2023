find winning numbers
first match is one and then every match after that is doubled
Ex:
Card 1: has 4 winning numbers so 1, 2, 4, 8

- Card 2 has two winning numbers (32 and 61), so it is worth 2 points.
- Card 3 has two winning numbers (1 and 21), so it is worth 2 points.
- Card 4 has one winning number (84), so it is worth 1 point.
- Card 5 has no winning numbers, so it is worth no points.
- Card 6 has no winning numbers, so it is worth no points.

```python
winning_dict = {}
scratch_nums_dict = {}

def get_winning_nums(row):
    card_num, winning_nums = row.split("|")[0].split(":")
    scratch_nums = row.split("|")[1]
    winning_dict[card_num] = winning_nums.split()
    scratch_nums_dict[card_num] = scratch_nums.split()
    return winning_dict,scratch_nums

for card in file:
    get_winning_nums(card)

matching_numbers = {}

for card, numbers in winning_dict.items():
    scratch_card_vals = scratch_nums_dict[card]
    matches = set(numbers) & set(scratch_card_vals)
    matching_numbers[card] = list(matches)

print(matching_numbers)

# get total score
total_score = 0
for card, matches in matching_numbers.items():
    if len(matches) == 0:
        total_score+=0
    else:
        score = 1
        for i in range(1, len(matches)):
            score *= 2
        total_score+= score

print(f"🚀{total_score}")
```
